```python
# The results of the calculations are added as new columns on the above table as shown below

# NDCI (Normalized Difference chlorophyl Index) --- (red edge-red)/(red edge + red)
sentinel_complete['NDCI'] = (sentinel_complete['Band5_red_edge']-sentinel_complete['Band4_red'])/(sentinel_complete['Band5_red_edge']+sentinel_complete['Band4_red'])


#salinity --- (Normalized Difference salinity Index) --- (Red-NIR)/(Red+NIR)
sentinel_complete['Water salinity'] = (sentinel_complete['Band4_red']-sentinel_complete['Band8_NIR'])/(sentinel_complete['Band4_red']+sentinel_complete['Band8_NIR'])


#ph ------ 8.339-0.827*(Coastal aerosol/NIR)
sentinel_complete['Water pH'] = 8.339-0.827*(sentinel_complete['Band1_aerosol']/sentinel_complete['Band8_NIR'])

sentinel_complete
```

# CLOUD MASKING  -- REMOVES THE CLOUDY PIXELS FROM AN IMAGE

```python
# Masking was necessary for landsat images since they have a longer revist period unlike sentinel hence fewer images taken per month therefore I needed to work with what I had
def maskL8sr(image): 
  #Bit 0 - Fill
  #Bit 1 - Dilated Cloud
  #Bit 2 - Cirrus
  #Bit 3 - Cloud
  #Bit 4 - Cloud Shadow
  qaMask = image.select('QA_PIXEL').bitwiseAnd(int('11111', 2)).eq(0);
  saturationMask = image.select('QA_RADSAT').eq(0);

  #Apply the scaling factors to the appropriate bands.
  opticalBands = image.select('SR_B.').multiply(0.0000275).add(-0.2)
  thermalBands = image.select('ST_B.*').multiply(0.00341802).add(149.0)

  #Replace the original bands with the scaled ones and apply the masks.
  return image.addBands(opticalBands, None, True).addBands(thermalBands, None, True).updateMask(qaMask).updateMask(saturationMask)


```

```python
# Map the masking function over the images from 2018 to 2022.
landsat = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2')\
                     .filterDate('2018-01-01', '2022-12-31')\
                     .filterBounds(area_of_interest)\
                     .map(maskL8sr)
```

```python
# calculating chlrophyll
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(8,25))
sentinel_complete_temp['Chlorophyll_content'] = scaler.fit_transform(sentinel_complete_temp['NDCI'].values[:, None])


sentinel_complete_temp
```
